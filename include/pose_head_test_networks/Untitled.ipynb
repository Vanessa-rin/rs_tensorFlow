```python
from PIL import Image
import json
import numpy as np
import tensorflow as tf

for i in [1,5,27,18,20]:
    net = tf.keras.models.load_model(f'test_inference_csl_{i}', custom_objects=None, compile=False, options=None)
#     net.summary()
    print('inputs:', net.inputs)
    print('outputs:', net.outputs)
    
    for j in range(5):
        print(i,j,':')
        
        rgb = np.array(Image.open(f'rgb_{i}_{j}.png'), np.float)
        depth = np.array(Image.open(f'depth_{i}_{j}.png'), np.float) / 10.
        with open(f'info_{i}_{j}.txt') as json_file:
            info = json.load(json_file)
            
            cam_K = np.array(info['cam_K']).reshape((3,3))
            coord_K = np.array(info['coord_K']).reshape((2,2))
            
            R = np.array(info['R']).reshape((3,3))
            t = np.array(info['t'])
            
        print(cam_K) 
        print(coord_K) 
        print(R) 
        print(t) 
        print('result:')
        res = net.predict([rgb[np.newaxis], depth[np.newaxis], cam_K[np.newaxis], coord_K[np.newaxis]])
        print(res[0][0])
        print(res[1][0])
        
        
```
